### Curvas de Indiferencia, Utilidad y Producci√≥n

### Curvas de Indiferencia y/o Utilidad

üîπ 1. Curvas de Utilidad
Definici√≥n: Representan el nivel de satisfacci√≥n (utilidad total) que un consumidor obtiene al consumir combinaciones de bienes.

Ejemplo: Si tienes 3 manzanas y 2 naranjas, y eso te da 100 ‚Äúutiles‚Äù de satisfacci√≥n, esa combinaci√≥n est√° en una curva de utilidad de 100.

### Librerias

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

# === 1. CURVAS DE INDIFERENCIA (UTILIDAD) ===
x = np.linspace(0.1, 10, 100)
y = np.linspace(0.1, 10, 100)
X, Y = np.meshgrid(x, y)
U = X**0.5 * Y**0.5
niveles_utilidad = [1, 2, 3, 4, 5]

plt.figure(figsize=(8, 6))
contours = plt.contour(X, Y, U, levels=niveles_utilidad, cmap="viridis")
plt.clabel(contours, inline=True, fontsize=10)
plt.title("Curvas de Indiferencia - Funci√≥n de Utilidad Cobb-Douglas")
plt.xlabel("Bien X")
plt.ylabel("Bien Y")
plt.grid(True)
plt.savefig("curvas_indiferencia_utilidad.png")
plt.show()

```

### Curvas de Producci√≥n (Isocuantas)

üîπ 3. **Curvas de Producci√≥n (Frontera de posibilidades de producci√≥n - FPP)**
- **Definici√≥n**: Representan las **combinaciones m√°ximas de dos bienes** que una econom√≠a puede producir utilizando todos sus recursos de manera eficiente.
- **Usadas por**: Empresas o econom√≠as, no consumidores.
- **Conceptos involucrados**: Eficiencia, escasez, costo de oportunidad.


```python
# === 2. CURVAS DE PRODUCCI√ìN (ISOCUANTAS) ===
L = np.linspace(0.1, 20, 100)
K = np.linspace(0.1, 20, 100)
L_grid, K_grid = np.meshgrid(L, K)
Q = L_grid**0.6 * K_grid**0.4
niveles_produccion = [5, 10, 15, 20, 25]

plt.figure(figsize=(8, 6))
contours = plt.contour(L_grid, K_grid, Q, levels=niveles_produccion, cmap="plasma")
plt.clabel(contours, inline=True, fontsize=10)
plt.title("Curvas de Producci√≥n (Isocuantas) - Cobb-Douglas")
plt.xlabel("Trabajo (L)")
plt.ylabel("Capital (K)")
plt.grid(True)
plt.savefig("curvas_produccion_isocuantas.png")
plt.show()

```

### Curva de Producci√≥n + Isocostes

```python
# === 3. CURVAS DE PRODUCCI√ìN + L√çNEA ISOCOSTE ===
C = 100   # presupuesto total
w = 5     # precio del trabajo
r = 10    # precio del capital
K_isocoste = (C - w * L) / r

plt.figure(figsize=(8, 6))
contours = plt.contour(L_grid, K_grid, Q, levels=niveles_produccion, cmap="plasma")
plt.clabel(contours, inline=True, fontsize=10)
plt.plot(L, K_isocoste, label="L√≠nea Isocoste", color="blue", linestyle="--")
plt.title("Curvas de Producci√≥n (Isocuantas) + L√≠nea Isocoste")
plt.xlabel("Trabajo (L)")
plt.ylabel("Capital (K)")
plt.legend()
plt.grid(True)
plt.ylim(0, 20)
plt.savefig("curvas_isocuantas_isocoste.png")
plt.show()
```

### ‚úÖ **Similitudes**

```python
 Aunque todas son **curvas econ√≥micas** y se usan para representar decisiones, tienen **diferentes objetivos** y se utilizan en **diferentes contextos**. Aqu√≠ van sus **definiciones**, **similitudes** y **diferencias**:

1. **Gr√°ficas en dos dimensiones**: todas usan dos ejes para representar combinaciones de dos variables (bienes o productos).
2. **Tienen forma curva** (usualmente convexas): reflejan principios como la utilidad marginal decreciente o el costo de oportunidad creciente.
3. **Representan elecciones** bajo restricciones (presupuesto o recursos).

---
```

### C√°lculo del √≥ptimo

```python
import numpy as np
import matplotlib.pyplot as plt

# --- Par√°metros iniciales ---
presupuesto = 100000
precio_x = 15.99  # Camisetas
precio_y = 49.99  # Zapatos
a = 0.5  # Preferencias iguales

# --- Punto √≥ptimo ---
x_opt = (a * presupuesto) / precio_x
y_opt = ((1 - a) * presupuesto) / precio_y
u_max = (x_opt**a) * (y_opt**(1 - a))  # M√°xima utilidad alcanzable

# Imprimir por si quieres verlo
print(f'Punto √≥ptimo: X = {x_opt:.2f}, Y = {y_opt:.2f}, U* = {u_max:.2f}')

# --- Ejes X e Y para el gr√°fico ---
x = np.linspace(1, x_opt * 1.2, 300)
y_presupuesto = (presupuesto / precio_y) - (precio_x / precio_y) * x

# --- Curvas de utilidad (de 0.1 a 1 del m√°ximo) ---
niveles_utilidad = np.linspace(u_max * 0.2, u_max, 5)

def curva_utilidad(u, x, a=0.5):
    return (u / (x ** a)) ** (1 / (1 - a))

# --- Graficar ---
plt.figure(figsize=(12, 8))

# Recta presupuestaria
plt.plot(x, y_presupuesto, color='red', label='Recta presupuestaria', linewidth=2)

# Curvas de utilidad
for u in niveles_utilidad:
    y_u = curva_utilidad(u, x, a)
    y_u = np.where(y_u > y_presupuesto.max() * 1.2, np.nan, y_u)  # filtrar extremos
    plt.plot(x, y_u, linestyle='--', alpha=0.7, label=f'U = {int(u)}')

# Punto √≥ptimo
plt.plot(x_opt, y_opt, 'bo', label='√ìptimo (m√°x utilidad)')

# Gr√°fico final
plt.title('Recta presupuestaria y curvas de utilidad realistas')
plt.xlabel('Camisetas')
plt.ylabel('Zapatos')
plt.xlim(0, x.max())
plt.ylim(0, max(y_presupuesto) + 20)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

### Diferentes tipos de funciones pero mismo planteamiento.

```python
En la producci√≥n, adem√°s de Cobb-Douglas, se usan funciones como Leontief, CES y lineales.

En la utilidad, adem√°s de Cobb-Douglas, existen funciones lineales, CES y de Leontief.

Cada tipo de funci√≥n es √∫til para modelar diferentes tipos de comportamientos y tecnolog√≠as de producci√≥n o preferencias de consumo.
```
